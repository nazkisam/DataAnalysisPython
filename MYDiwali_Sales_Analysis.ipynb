```python
#plotting the graphs

grp = sns.countplot(x='Gender',data=df)

for bars in grp.containers:
  grp.bar_label(bars)
```

```python
grp = sns.countplot(data = df, x = 'Age Group', hue = 'Gender')
for bars in grp.containers:
  grp.bar_label(bars)
```

```python
df
```

```python
sales_age = df.groupby(['Age Group'], as_index = False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x='Age Group',y='Amount',data=sales_age)
```

```python
sales_state = df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set_theme(rc={'figure.figsize':(16,5)})
sns.barplot(data=sales_state,x='State',y='Orders')
```

```python
#total amount and sales from top 10 states

data_amount = df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set_theme(rc={'figure.figsize':(16,5)})

sns.barplot(data=data_amount,x='State',y='Amount')
```

```python
#count the marital status of the buyers, where 0 is single and 1 is married.

marital_status= sns.countplot(data = df, x=  'Marital_Status')
sns.set_theme(rc={'figure.figsize':(7,5)})

for bars in marital_status.containers:
 marital_status.bar_label(bars)
```

```python
maritalSales = df.groupby(['Marital_Status','Gender'], as_index = False)['Amount'].sum().sort_values(by='Amount',
                                                                                                     ascending = False)
sns.set_theme(rc={'figure.figsize':(7,5)})

sns.barplot(data = maritalSales,x= 'Marital_Status',y='Amount', hue = 'Gender')
```

```python
occupation = sns.countplot(data = df, x = 'Occupation')
sns.set_theme(rc={'figure.figsize':(30,5)})

for bars in occupation.containers:
  occupation.bar_label(bars)
```

```python
occupationPurchase = df.groupby('Occupation', as_index = False)['Amount'].sum().sort_values(by='Amount', 
                                                                                            ascending = False )
sns.set_theme(rc={'figure.figsize':(18,5)})

sns.barplot(data = occupationPurchase, x = 'Occupation', y = 'Amount')
```

```python
productCate = sns.countplot(data = df, x = 'Product_Category')

sns.set_theme(rc={'figure.figsize':(30,5)})

for bars in productCate.containers:
  productCate.bar_label(bars)
```

```python
soldCate = df.groupby(['Product_Category'], as_index = False)['Amount'].sum().sort_values(by='Amount', 
                                                                                         ascending = False).head(10)
sns.set_theme(rc= {'figure.figsize':(20,5)})

sns.barplot(data = soldCate, x = 'Product_Category', y= 'Amount')
```

```python
soldProduct = df.groupby('Product_ID', as_index = False)['Orders'].sum().sort_values(by = 'Orders',
                                                                                     ascending = False).head(10)
sns.set_theme(rc={'figure.figsize':(18,5)})

sns.barplot(data = soldProduct, x= 'Product_ID' , y = 'Orders')
```
